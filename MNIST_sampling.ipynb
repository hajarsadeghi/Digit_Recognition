```python
#Data Normalization

#convert to float
x_train = x_train.astype('float32')/255.
x_test = x_test.astype('float32')/255.


#reshape the dataset so that 28*28 dimention becomes 784, Convert each image to 1d array (28x28 -> 784x1)
x_train = x_train.reshape(x_train.shape[0], -1)
x_test = x_test.reshape(x_test.shape[0], -1)
```

# Question 1 - part 1

We know that the kmeans.labels_ only denotes the cluster to which the image belongs to. It doesn’t denote the number displayed in image. Hence we write a separate function to retrieve the necessary information from kmeans.labels_

We:

* Check Frequency distribution of actual labels (using np.bincount)
* Find the Maximum frequent label (through np.argmax), and set the label.

```python
#we will define a function that can give us the values associated with the kmeans labels
def get_image_info(kmodel, actual_labels):
  reference_labels = {}
  # print(kmodel.labels_.shape, 'labels shape')
  # print(actual_labels.shape, 'actual labels shape')
  for i in range(len(np.unique(kmodel.labels_))):
    index = np.where(kmodel.labels_ == i, 1, 0)
    num = np.bincount(actual_labels[index==1]).argmax()
    reference_labels[i] = num 

  return reference_labels
```

Instead of using the original Kmeans, we chose MiniBatchKMeans. The distinction is that with MiniBatchKMeans, the model does not need to use the entire dataset for each iteration; instead, the algorithm can employ mini-batches and only move the centroids slightly. This speeds up the method by a factor of three or four, allowing it to cluster large datasets that would otherwise be impossible to store in memory. So, since our dataset (MNIST) is a huge dataset, we use MiniBatchKMeans in our model.

Before training our model, we need to try different values of k to find the best parameter for our model. 
We are going to use the elbow method, and silhouette score, to determine the best value for k.

* The elbow method uses the sum of squared distance (SSE) (inertia_ in sklearn) to choose an ideal value of k based on the distance between the data points and their assigned clusters. We would choose a value of k where the SSE begins to flatten out and we see an inflection point.
Elbow method says that you can find K where there is an elbow, i.e. before elbow error gradually decreases and after elbow the decrease isn’t significant.

* A more precise approach (but also more computationally expensive) is to use the silhouette score, which is the mean silhouette coefficient over all the instances. An instance’s silhouette coefficient is equal to (b – a) / max(a, b) where a is the mean distance to the other instances in the same cluster (it is the mean intra-cluster distance), and b is the mean nearest-cluster distance, that is the mean distance to the instances of the next closest cluster (defined as the one that minimizes b, excluding the instance’s own cluster). The silhouette coefficient can vary between -1 and +1: a coefficient close to +1 means that the instance is well inside its own cluster and far from other clusters, while a coefficient close to 0 means that it is close to a cluster boundary, and finally a coefficient close to -1 means that the instance may have been assigned to the wrong cluster.

```python
from sklearn.cluster import MiniBatchKMeans
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import silhouette_samples, silhouette_score
from matplotlib import markers
from yellowbrick.cluster import SilhouetteVisualizer
import numpy as np

cluster_array = [10, 30, 50, 70, 90, 110, 130, 200, 300]
distortions = []
silu = []
for i in cluster_array:
  # Initialize the K-Means model
  kmeans = MiniBatchKMeans(n_clusters = i, init='random', random_state=0)
  # Fitting the model to training set
  kmeans.fit(x_train)
  distortions.append(kmeans.inertia_)
  silu_i = silhouette_score(x_train, kmeans.labels_)
  silu.append(silu_i)

  # # Calculating cost
  print('Number of clusters is {}'.format(i))
  print('Inertia : {}'.format(kmeans.inertia_))

  # Calculating reference_labels
  reference_labels = get_image_info(kmeans,y_train)
  # ‘labels_mat’ is a list which denotes the number displayed in image
  labels_mat = np.random.rand(len(kmeans.labels_))

  for j in range(len(kmeans.labels_)):
    labels_mat[j] = reference_labels[kmeans.labels_[j]]

  visualizer = SilhouetteVisualizer(kmeans, colors='yellowbrick')

  visualizer.fit(x_train)        # Fit the data to the visualizer
  visualizer.show()        # Finalize and render the figure

  print('Accuracy score : {}'.format(accuracy_score(labels_mat,y_train)))
  print('Silhouette score : {}'.format( silu_i ))
  print('\n')

  %time
```

```python
#plot the distortions for the elbow method
plt.plot(cluster_array, distortions, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Distortion')
plt.tight_layout()
plt.show()

#plot the Silhouette score for different clusters
plt.plot(cluster_array, silu, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette score')
plt.tight_layout()
plt.show()
```

In the above, we first plotted the silhouette diagrams for different values of k. Then, we plotted the distortion and silhouette scores for each of them. 
First, based on the elbow method (distortion plot), we see that after the 70 and 90 clusters (the fourth and fifth points), the distortion value does not decrease significantly, so each of them can be a good indicator for the value of k. 
Moreover, based on the silhouette score plot, we see that in the 70 cluster, we have a peak indicating that choosing 70 as our cluster can be a reasonable choice. Apart from 70, the 10 and 90 clusters also have high silhouette scores, but the accuracy for 10 is around 50% which is really low, but choosing 90 can be efficient because the accuracy improves to 83%.
Therefore, we will take a closer look at the silhouette diagrams to choose between 70 and 90 as the value of k.
We see that, for 90, clusters are more similar in size compared to clusters in 70, so we choose 90 for our cluster size.

# **Question 1 - part 4: Manually initializing cluster center**

First, we randomly choose 1 sample from each of the 10(0 to 9) classes to serve as the initial cluster centers.

```python
from numpy.random.mtrand import sample
import numpy as np

n_clusters = 10
row = []
initial_centroid = []
for i in range(n_clusters):
    #indices of all the points from X that belong to cluster i
    C_i = np.where(y_train == i)[0].tolist() 
    #indices of the points from X to be sampled from cluster i
    sample_i = np.random.choice(C_i, int(1))
    temp = np.asarray(x_train[sample_i])
    row.append(sample_i)
    initial_centroid.append(temp)


initial_centroid = np.asarray(initial_centroid)
initial_centroid = initial_centroid.reshape(initial_centroid.shape[0], -1)

#Visualize the initial centroids
plt.figure(figsize=(20, 25))
for i in range(n_clusters):
  j = row[i]
  plt.subplot(16, 16, i+1)
  plt.title('Number:{}'.format(y_train[j]),fontsize = 10)
  plt.xticks([])
  plt.yticks([])
  plt.imshow(initial_centroid[i].reshape(28, 28)*255)
```

```python
from datetime import datetime
start_time = datetime.now()

k_cluster = 10

print("Intialized by us")
kmeans_intialized = MiniBatchKMeans(n_clusters = k_cluster, init=initial_centroid, random_state=0, n_init=10)
# Fitting the model to training set
kmeans_intialized.fit(x_train)
# Calculating reference_labels
reference_labels = get_image_info(kmeans_intialized,y_train)
# ‘labels_mat’ is a list which denotes the predicted number displayed in image
labels_mat = np.random.rand(len(kmeans_intialized.labels_))

for j in range(len(kmeans_intialized.labels_)):
  labels_mat[j] = reference_labels[kmeans_intialized.labels_[j]]

print('Accuracy score : {}'.format(accuracy_score(labels_mat,y_train)))
print('Number of clusters is {}'.format(k_cluster))
print('Inertia : {}'.format(kmeans_intialized.inertia_))
print('\n')
end_time = datetime.now()
print('Duration: {}'.format(end_time - start_time))
# %time
start_time = datetime.now()

print('-'*20)
start_time = datetime.now()
print("Random initialization")
kmeans_rd = MiniBatchKMeans(n_clusters = k_cluster, init='random', random_state=0, n_init=10)
# Fitting the model to training set
kmeans_rd.fit(x_train)
# Calculating reference_labels
reference_labels = get_image_info(kmeans_rd,y_train)
# ‘labels_mat’ is a list which denotes the predicted number displayed in image
labels_mat = np.random.rand(len(kmeans_rd.labels_))

for j in range(len(kmeans_rd.labels_)):
  labels_mat[j] = reference_labels[kmeans_rd.labels_[j]]

print('Accuracy score : {}'.format(accuracy_score(labels_mat,y_train)))
print('Number of clusters is {}'.format(k_cluster))
print('Inertia : {}'.format(kmeans_rd.inertia_))
print('\n')
end_time = datetime.now()
print('Duration: {}'.format(end_time - start_time))
# %time
start_time = datetime.now()

print('-'*20)
start_time = datetime.now()
print("k-means++ initialization")
kmeans_plus = MiniBatchKMeans(n_clusters = k_cluster, init='k-means++', random_state=0, n_init=10)
# Fitting the model to training set
kmeans_plus.fit(x_train)
# Calculating reference_labels
reference_labels = get_image_info(kmeans_plus,y_train)
# ‘labels_mat’ is a list which denotes the predicted number displayed in image
labels_mat = np.random.rand(len(kmeans_plus.labels_))

for j in range(len(kmeans_plus.labels_)):
  labels_mat[j] = reference_labels[kmeans_plus.labels_[j]]

print('Accuracy score : {}'.format(accuracy_score(labels_mat,y_train)))
print('Number of clusters is {}'.format(k_cluster))
print('Inertia : {}'.format(kmeans_plus.inertia_))
print('\n')
end_time = datetime.now()
print('Duration: {}'.format(end_time - start_time))
```

As we can see, the computation time for **customized inialization** for values of centroids is slightly better than 'Random' or 'K-Means++'. We think, the reason is that we chose the initial centroids based on each digit from 0 to 9, so the model converges faster.
To make sure about this concolusion, we ran the code above several times and each time we got about the same results.

# **Question 1 - part 6: Comparing the variance in the above three approaches**

```python
# Random
print("----- Random -----")
variance = 0
for ci,centroid in enumerate(centroids_random):
  for cluster_point in x_test[np.where(clusters_random == ci)[0]]:
     variance += np.linalg.norm(cluster_point - np.reshape(centroid, (784,)))
print(variance)
print()

# Kmeans++
print("----- Kmeans++ -----")
variance = 0
for ci,centroid in enumerate(centroids_kplus):
  for cluster_point in x_test[np.where(clusters_kplus== ci)[0]]:
     variance += np.linalg.norm(cluster_point - np.reshape(centroid, (784,)))
print(variance)
print()

#Initialized
print("----- Initialized -----")
variance = 0
for ci,centroid in enumerate(centroids_initial):
  for cluster_point in x_test[np.where(clusters_initial == ci)[0]]:
     variance += np.linalg.norm(cluster_point - np.reshape(centroid, (784,)))
print(variance)
print()
```

In the first two models we had 90 clusters and in the last one we had 10 clusters. We can see that the more clusters we have, the higher would be the variance. The reason is that imagine we have just one cluster, then the centroid would be the actual center of the whole dataset. Thus, it makes sense that as we add more clusters, the number of data points that could be added to a cluster is decraseing.
